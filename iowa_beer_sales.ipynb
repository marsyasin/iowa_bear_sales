```python
#1. Calculate the profit_margin for each sales

import numpy as np

# Calculate the profit_margin
df['profit_margin'] = (((df['sale_price_dollars'] - (df['state_bottle_cost'] * df['bottles_sold'])) / 
                        df['sale_price_dollars']) * 100)


df['profit_margin'].replace([np.inf, -np.inf], np.nan, inplace=True)
df['profit_margin'].fillna(0, inplace=True)

df['profit_margin'] = df['profit_margin'].round(2)

#2. Calculate the average profit margin by using mean() method
                        
average_margin_percentage = df['profit_margin'].mean()
print('\nAverage Profit Margin Percentage:', average_margin_percentage)
```

```python
#1. Based on the DataFrame head, I will look for the values within category to look for the liquor types

unique_category = df['category_name'].unique()
unique_category_list = unique_category.tolist()
print(unique_category_list)
```

```python
#2. To show the sales of each liquor type, I will create new DataFrames for each liquor type

# Create a function to filter, group, and count the sales
# The function will create a new DataFrame for each liquor type by filtering with keywords
# Each DataFrame will be grouped by each category name
# Sales will be counted based on each group of the brand names

def calculate_sales_by_category(df, keyword):
    filtered_df = df[df['category_name'].str.contains(keyword, case=False)]
    grouped_category = filtered_df.groupby('category_name')
    return grouped_category['sale_price_dollars'].sum()


# A list for the liquor type and the sales for each liquor type
liquor_types = ['vodka', 'whiskies', 'cordial']
sales_by_category = {liquor_type: calculate_sales_by_category(df, liquor_type) for 
                     liquor_type in liquor_types}

# Sort the sales list from highest to lowest because we are looking for the top sales
sorted_sales_by_category = {liquor_type: sales_by_category[liquor_type].sort_values(ascending=False) 
                            for liquor_type in liquor_types}

# Plot bar charts for each liquor type for an easier analysis
for liquor_type, sales in sorted_sales_by_category.items():
    plt.figure(figsize=(10, 6))
    plt.barh(sales.index, sales.values, color='skyblue')
    plt.xlabel('Total Sales (USD)')
    plt.ylabel('Category Name')
    plt.title(f'Total Sales of {liquor_type}')
    plt.gca().invert_yaxis()
    plt.show()
```

```python
top_5_categories = df['category_name'].value_counts().nlargest(5).index.tolist()

filtered_df = df[df['category_name'].isin(top_5_categories)]
```

```python
X = filtered_df[['bottle_volume', 'state_bottle_cost', 'volume_sold_liters']] 
y = filtered_df['category_name']
```

```python
from sklearn.tree import DecisionTreeClassifier
dummy_df = pd.get_dummies(filtered_df['category_name'])

X_classification = pd.concat([filtered_df[['bottle_volume', 'state_bottle_cost', 'volume_sold_liters']], dummy_df], axis=1)
y_classification = filtered_df['category_name']

assert X_classification.shape[0] == y_classification.shape[0], "Number of samples in X and y do not match"

X_train, X_test, y_train, y_test = train_test_split(X_classification, y_classification, test_size=0.25, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

dt = DecisionTreeClassifier(random_state=1)
dt.fit(X_train_scaled, y_train)

y_pred = dt.predict(X_test_scaled)
y_train_pred = dt.predict(X_train_scaled)

# Calculate accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_pred)

print('Train Accuracy:', train_accuracy)
print('Test Accuracy:', test_accuracy)
```
